```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ResNet50 = pd.read_json("../history/CIFAR10-ResNet50-TRAINHISTORY.json")
AlexNet = pd.read_json("../history/CIFAR10-AlexNet-TRAINHISTORY.json")
VGGNet16 = pd.read_json("../history/CIFAR10-VGGNet16-TRAINHISTORY.json")
```

```python
models = [("ResNet50",ResNet50), ("AlexNet", AlexNet), ("VGGNet16", VGGNet16)]
for name, model in models:
    plt.figure(figsize=(8, 6))

    plt.subplot(211)
    plt.plot(model["Test Accuracy"])
    plt.plot(model["Train Accuracy"])
    plt.legend(["Test Accuracy", "Train Accuracy"])
    plt.grid(True)
    plt.title(name)

    plt.subplot(212)
    plt.plot(model["Test Loss"])
    plt.plot(model["Train Loss"])
    plt.legend(["Test Loss", "Train Loss"])

    plt.grid(True)
    plt.show()
```

```python
ResNet50 = pd.read_json("../history/Melasema-ResNet50-TRAINHISTORY.json")
AlexNet = pd.read_json("../history/Melasema-AlexNet-TRAINHISTORY.json")
VGGNet16 = pd.read_json("../history/Melasema-VGGNet16-TRAINHISTORY.json")
```
